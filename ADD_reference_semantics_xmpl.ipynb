### Addendum: Memory

#### Reference semantics

This examples highlights the main differences between a shallow copy and a deep copy.

```python
inner_list = [1,2]
lst0 = [inner_list, 3]
```

```python
lst1 = lst0
```

```python
lst2 = lst0.copy() # shallow  copy
```

```python
from copy import deepcopy
lst3 = deepcopy(lst0)
```

```python
lst0[1] = 'three'
print(lst1)
print(lst2)
print(lst3)
```

```python
lst0[0][0] = 'one'
print(lst1)
print(lst2)
print(lst3)
```
