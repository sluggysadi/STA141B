```python
monomials = [lambda x: x**i for i in range(5)]
```

```python
monomials[2](3) # does not return 3**2!
```

```python
monomials[0](3)
```

The reason for this is that the i inside the lambda function will be evaluated after the for loop finished. So all monomials will be the same function ($f(x) = x^4$)

To solve this:

```python
monomials = [lambda x, l=i: x**l for i in range(5)]
```

```python
monomials[2](6)
```

```python
[f(3) for f in monomials]
```
